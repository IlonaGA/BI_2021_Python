Проведите Explorative Data Analysis какого-нибудь датасета (прошу, только не Титаник). Нужны корреляции, графики распределений

```python
marks_df = pd.read_excel('marks.xlsx', sheet_name='Python')
marks_df.head()
```

```python
names = marks_df.columns[7:]
```

```python
marks_df.loc[:, names].fillna(0, inplace=True)
```

```python
for name in names:
    plt.hist(marks_df[name], bins=15, range=[0, marks_df[name].max() + 1])
    plt.title(name)
    plt.show()
```

```python
marks_df[names].mean()
```

```python
marks_df[names].median()
```

```python
plt.figure(figsize=(16, 9))
plt.title('Marks correlation')
sns.heatmap(marks_df[names].corr(), vmin=-1, vmax=1, annot=True)
plt.show()
```
